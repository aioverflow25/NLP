```python
corpus = """Hello welcome, to Raman Singh's NLP Course.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
import nltk
nltk.download('punkt_tab')
## sent_tokenize() internally uses the “punkt” tokenizer model, which isn’t downloaded automatically when you install NLTK.
```

```python
# Tokenization
# Corpus -> Documents
# Paragraph -> Sentances
from nltk.tokenize import sent_tokenize
sentances = sent_tokenize(corpus)
for i,sent in enumerate(sentances,1):
    print(f"{i}. {sent}")
```

```python
## Word TOkenize
## Documents -> Words
## Paragraph/sentances -> WOrds
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentance in sentances:
    print(word_tokenize(sentance))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
## Punctuations also get splitted 
## "'s was a single token now "'" "s" are 2 tokens
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer = TreebankWordTokenizer()
tokenizer.tokenize(corpus)
## With respect to last full stop it will considerd as different word but for all other full stop it will not consider it as diff word
```
