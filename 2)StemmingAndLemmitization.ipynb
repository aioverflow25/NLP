```python
words = [
    "running", "runner", "runs",
    "easily", "fairly",
    "flying", "flies", "flight",
    "studies", "studying", "studied",
    "playing", "played", "plays",
    "happily", "happiness", "happy",
    "argues", "arguing", "argument",
    "connection", "connected", "connecting",
    "relational", "relate", "relation"
]
```

```python
## Porter Stemmer
from nltk.stem import PorterStemmer
stmr = PorterStemmer()
for word in words:
    print(word+"---->"+stmr.stem(word))
```
