```python
corpus = """
John loves playing football with his friends every Sunday afternoon. 
He runs quickly across the field and scores amazing goals. 
Sometimes, the weather becomes cloudy, but they still continue to play happily. 
After the match, everyone feels tired yet satisfied because teamwork always brings joy.
"""
import nltk
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
## Pos tagging
import nltk
from nltk.tokenize import sent_tokenize
sentances = sent_tokenize(corpus)
for sent in sentances:
    words = nltk.word_tokenize(sent)
    pos_tag = nltk.pos_tag(words)
    print(pos_tag)
```

```python
sent = ["TajMahal is a beautiful Monument","Raman is Data Analyst","Gaduly is my posting"]
# We have to tokenize each sentance
tokenize_sentances = [nltk.word_tokenize(s) for s in sent]
pos_tag = nltk.pos_tag_sents(tokenize_sentances)
print(pos_tag)
```
